```python
from sklearn.model_selection import train_test_split 
import lightgbm as lgb 
from sklearn.linear_model import  LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_curve,auc 

labels = np.array(labels)
#x_train,x_test,y_train,y_test = x_std,x_std[-1000:,:],labels,labels[-1000:]
x_train,x_test,y_train,y_test = train_test_split(x_std,labels,train_size=0.9)
plt.figure(figsize=(8, 8))

logm = LogisticRegression()
rdfm = RandomForestRegressor(max_depth=12,random_state=0)
lgbm = lgb.LGBMRegressor(
    learning_rate=0.1,
    n_estimators=40
)
knn = KNeighborsRegressor(n_neighbors=10)

models = [logm,rdfm,lgbm,knn]
models_names = ['Logist',"RandomForest","Lgbm","KNN"]
for i,n in enumerate(models):
    m = models[i]
    m.fit(x_train,y_train)
    
    y_p = None
    if models_names[i] == 'Logist':
        y_p = m.predict_proba(x_test)[:,1]
    else:
        y_p = m.predict(x_test)

    fpr, tpr, thresholds = roc_curve(y_test, y_p)
    roc_auc = auc(fpr, tpr)
# 计算AUC
# 绘制ROC曲线
    plt.plot(fpr, tpr, lw=2, label=f'{models_names[i]} AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.legend(loc='lower right',fontsize = 10)
plt.xlabel('False Positive Rate',fontsize = 10)
plt.ylabel('True Positive Rate',fontsize = 10)
plt.title(' (ROC) Curve',fontsize = 10)
plt.show()

```

```python
from sklearn.model_selection import KFold 
from sklearn.metrics import mean_squared_error
nk = 20
kf = KFold(n_splits=nk)
sco_tr = []
sco_te = []

for k,(train_idx,test_idx) in enumerate(kf.split(x_std,labels)):
    
    x_train,x_test,y_train,y_test = x_std[train_idx],x_std[test_idx],labels[train_idx],labels[test_idx]
    m = RandomForestRegressor(max_depth=12,random_state=0)
    #m = lgb.LGBMRegressor(
    #learning_rate=0.01,
   # n_estimators=40
    #)
    m.fit(x_train,y_train)
    score_train = mean_squared_error(y_train,m.predict(x_train))
    score_test = mean_squared_error(y_test,m.predict(x_test))
    sco_tr.append(score_train)
    sco_te.append(score_test)
    print(f"{k}fold:",f"train:{sco_tr}",f"test:{sco_te}")


plt.figure(figsize=(8, 8))
  
plt.plot(np.arange(nk),sco_tr,label='train')
plt.plot(np.arange(nk),sco_te,label='test')

plt.legend(loc='lower right',fontsize = 10)
plt.xlabel('K times',fontsize = 10)
plt.ylabel('mean_squared_error',fontsize = 10)
plt.title('K-fold-val',fontsize = 10)
plt.show()
```
