```python
from sklearn.model_selection import train_test_split 
import lightgbm as lgb 
from sklearn.linear_model import  LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_curve,auc 

#labels = np.array(labels)
x_train,x_test,y_train,y_test = xo_std,xo_std[-400:,:],olabels,olabels[-400:]
#x_train,x_test,y_train,y_test = train_test_split(xo_std,olabels,train_size=0.9)
#x_train ,x_test,y_train,y_test = x_std,xw_std,np.array(labels),np.array(wlabels)
plt.figure(figsize=(8, 8))

logm = LogisticRegression()
rdfm = RandomForestRegressor(random_state=0)
lgbm = lgb.LGBMRegressor(
    learning_rate=0.1,
    n_estimators=40
)
knn = KNeighborsRegressor(n_neighbors=10)

models = [logm,rdfm,lgbm,knn]
models_names = ['Logist',"RandomForest","Lgbm","KNN"]
for i,n in enumerate(models):
    m = models[i]
    m.fit(x_train,y_train)
    
    y_p = None
    if models_names[i] == 'Logist':
        y_p = m.predict_proba(x_test)[:,1]
    else:
        y_p = m.predict(x_test)

    fpr, tpr, thresholds = roc_curve(y_test, y_p)
    roc_auc = auc(fpr, tpr)
# AUC
# ROC
    plt.plot(fpr, tpr, lw=2, label=f'{models_names[i]} AUC = {roc_auc:.2f}')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.legend(loc='lower right',fontsize = 10)
plt.xlabel('False Positive Rate',fontsize = 10)
plt.ylabel('True Positive Rate',fontsize = 10)
plt.title(' (ROC) test on men',fontsize = 10)
plt.show()

```
